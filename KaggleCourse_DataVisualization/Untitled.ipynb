## Importing necessary libraries

## 1) Exploring the FIFA Dataset

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Read the file into a variable fifa_data
fifa_data = pd.read_csv("data/fifa.csv", index_col="Date", parse_dates=True)
```

## Examining the dataset

```python
fifa_data.head()
```

```python
plt.figure(figsize=(16,6))

sns.lineplot(data=fifa_data)

plt.title("Evolution of FIFA rankings over time")
plt.ylabel("Ranking")
plt.show()
```

## 3) Exploring Flight Delay Dataset

## 4) Exploring Insurance Dataset

```python
ins_data = pd.read_csv("data/insurance.csv")
```

```python
ins_data.head()
```

```python
#finding correlation between bmi and charges

sns.scatterplot(x=ins_data['bmi'], y=ins_data['charges'])

plt.title("Relation between Insurance Charges and BMI")
plt.show()
```

```python
sns.regplot(x=ins_data['bmi'], y=ins_data['charges'])

plt.title("Relation between Insurance Charges and BMI")
plt.show()
```

### Color-coded scatter plots


```python
sns.scatterplot(x=ins_data['bmi'], y=ins_data['charges'], hue=ins_data['smoker'])

plt.title("Relationship between BMI, Insurance Costs & Smoking")
plt.show()
```

```python
#Adding two regression lines, corresponding to smokers and nonsmokers
sns.lmplot(x="bmi", y="charges", hue="smoker", data=ins_data)

plt.show()
```

```python
sns.swarmplot(x=ins_data['smoker'], y=ins_data['charges'])

plt.show()
```

## 5) Exploring the IRIS Dataset

```python
iris_data = pd.read_csv("data/iris.csv", index_col="Id")
```

```python
iris_data.head()
```

```python
# Histogram 
sns.distplot(a=iris_data['Petal Length (cm)'], kde=False)

plt.show()
```

### Density plots

```python
# KDE plot 
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)

plt.show()
```

### 2D KDE plots

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")

plt.show()
```

### Color-coded plots

```python
# Paths of the files to read
iris_set_filepath = "data/iris_setosa.csv"
iris_ver_filepath = "data/iris_versicolor.csv"
iris_vir_filepath = "data/iris_virginica.csv"

# Read the files into variables 
iris_set_data = pd.read_csv(iris_set_filepath, index_col="Id")
iris_ver_data = pd.read_csv(iris_ver_filepath, index_col="Id")
iris_vir_data = pd.read_csv(iris_vir_filepath, index_col="Id")

# Print the first 5 rows of the Iris versicolor data
iris_ver_data.head()
```

```python
# Histograms for each species
sns.distplot(a=iris_set_data['Petal Length (cm)'], label="Iris-setosa", kde=False)
sns.distplot(a=iris_ver_data['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=iris_vir_data['Petal Length (cm)'], label="Iris-virginica", kde=False)

plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
plt.show()
```

```python
# KDE plots for each species
sns.kdeplot(data=iris_set_data['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_ver_data['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_vir_data['Petal Length (cm)'], label="Iris-virginica", shade=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
plt.show()
```
